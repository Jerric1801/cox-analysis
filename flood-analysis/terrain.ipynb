## Run-Off Equation

$$Q = \begin{cases} 
  0 & \text{for } P \leq I_a \\
  \frac{(P - I_a)^2}{P - I_a + S} & \text{for } P > I_a 
  \end{cases}$$

where

- Q is runoff
- P is rainfall
- S is potential maximum soil moisture retention after runoff begins
- $I_a$  is initial abstraction

### Inches 
$$S = \frac{1000}{CN} - 10$$ 

### cm
$$S = \frac{25400}{CN} - 25.4$$
